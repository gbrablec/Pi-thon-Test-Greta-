# How to Find $\pi$ with Python


This notebook gives an example of how mathematical questions can be answered by a few lines of code. Here, we want to estimate $\pi$ by making a computer throw darts!

If you throw darts on something that looks like the figure above, the proability of landing inside the circle is $\frac{\pi}{4}$. If you throw a lot of darts, the fraction of darts inside the circle will be $\approx$ $\frac{\pi}{4}$. We can find $\pi$ by throwing darts then, which is an example of the *Monte Carlo method*. <br><br> **Here are some resources**:
 - [Monte Carlo method](https://en.wikipedia.org/wiki/Monte_Carlo_method)
 - [How to Throw Like a Pro!](youtube.com/watch?v=9gB5ybp_7s8)

Before doing things in Python, you usually need to import a few libraries. These libraries have special commands. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

We want to throw darts at a cirlce, so we need to create the circle using equation:<br>$x^2+y^2=r^2$<br>You can also write x as $r\cos\theta$ and y as $r\sin\theta$ where $\theta$ is the angle at which (x,y) is located. Now, we will select a lot of angles ($\theta$ values) between 0 and 2$\pi$. This will give a lot of (x,y) points.

```python
theta = np.linspace(0,2*np.pi,100)
x = np.cos(theta)
y = np.sin(theta)
```

x is a list of x coordinates.

```python
plt.plot(x,y)
```

Now, we need to inscribe the circle in a square. To do this, we must draw 4 equal lines. The following code draws a purple line from (-1,1) to (-1,-1).

```python
plt.plot([-1,-1],[1,-1],'m')
```

Time to inscribe!

```python
plt.plot(x,y)
plt.plot([-1,-1],[1,-1], 'm')
plt.plot([1,1],[1,-1], 'm')
plt.plot([-1,1],[-1,-1], 'm')
plt.plot([1,-1],[1,1], 'm')
```

To throw a dart, we make Python randomly choose a point (x,y). We achieve this by randomly choosing a number between -1 and 1 and assigning that value to both x and y.

```python
plt.plot(x,y)
plt.plot([-1,-1],[1,-1], 'm')
plt.plot([1,1],[1,-1], 'm')
plt.plot([-1,1],[-1,-1], 'm')
plt.plot([1,-1],[1,1], 'm')
a = np.random.uniform(-1,1,1)
b = np.random.uniform(-1,1,1)
plt.plot(a,b, 'k.')
```

If we run the above cell a few time, it'll be a different point each time.<br>To check if the random point is inside of the circle, it must satisfy the inequality $x^2+y^2<r^2$.

```python
if a**2+b**2 <=1:
    print("It's inside!")
else:
    print("Oh No!")
```

We need to throw many darts and count how many are inside of the circle. We do this using a list, and we start with an empty list.

We start with two empty list values of a and b respectively.

```python
avec =[]
bvec =[]
```

We set ntotal to be the number of darts we want to throw. We also define ninside as the number of darts inside the circle. It starts out at 0, but every time a dart falls inside the circle, we will increase it by 1.

```python
ntotal =1000
ninside =0
```

```python
i = 0

while i <=ntotal:
    a = np.random.uniform(-1,1,1)
    b = np.random.uniform(-1,1,1)
    if a**2+b**2 <=1:
        ninside = ninside +1
    avec = avec +[a]
    bvec = bvec +[b]
    i +=1
```

```python
plt.plot(x,y)
plt.plot([-1,-1],[1,-1],'m')
plt.plot([1,1],[1,-1],'m')
plt.plot([-1,1],[-1,-1],'m')
plt.plot([1,-1],[1,1],'m')
plt.plot(avec,bvec,'k.')
```

```python
print(4*ninside/ntotal)
```
